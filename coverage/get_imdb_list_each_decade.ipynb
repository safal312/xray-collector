```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files

def scrape_movies(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    movies = soup.find_all('div', class_='lister-item mode-detail')
    movie_titles = [movie.find('h3', class_='lister-item-header').a.text.strip() for movie in movies]
    movie_hrefs = [movie.find('h3', class_='lister-item-header').a['href'] for movie in movies]
    return movie_titles, movie_hrefs

def main(url, decade):
    movie_list, movie_links = scrape_movies(url)
    filename = f'{decade}_movie_list.csv'
    df = pd.DataFrame(movie_list, columns=['Movie Title'])
    df['decade'] = decade
    df['id'] = movie_links
    df['id'] = df['id'].str.extract("(tt[0-9]+)")
    return df
    # files.download(filename)  # Download the file in Colab
```

```python
all_dfs = []

# if __name__ == "__main__":
LINKS = ['https://www.imdb.com/list/ls026603988/', 'https://www.imdb.com/list/ls026659297/','https://www.imdb.com/list/ls000050037/','https://www.imdb.com/list/ls026673591/','https://www.imdb.com/list/ls042211303/','https://www.imdb.com/list/ls070213744/','https://www.imdb.com/list/ls000098551/','https://www.imdb.com/list/ls000098551/','https://www.imdb.com/list/ls000064427/','https://www.imdb.com/list/ls002065120/','https://www.imdb.com/list/ls058598700/','https://www.imdb.com/list/ls096194972/']
for index, decade in enumerate(['1910s','1920s','1930s','1940s','1950s','1960s','1970s','1980s','1990s', '2000s', '2010s', '2020s']):
    df = main(LINKS[index], decade)
    all_dfs.append(df)
    # main('https://www.imdb.com/list/ls026659297/', '1920s')
    # main('https://www.imdb.com/list/ls000050037/', '1930s')
    # main('https://www.imdb.com/list/ls026673591/', '1940s')
    # main('https://www.imdb.com/list/ls042211303/', '1950s')
    # main('https://www.imdb.com/list/ls070213744/', '1960s')
    # main('https://www.imdb.com/list/ls000098551/', '1970s')
    # # main('https://www.imdb.com/list/ls000098551/', '1980s')
    # # main('https://www.imdb.com/list/ls000064427/', '1990s')
    # # main('https://www.imdb.com/list/ls002065120/', '2000s')
    # # main('https://www.imdb.com/list/ls058598700/', '2010s')
    # # main('https://www.imdb.com/list/ls096194972/', '2020s')
```
