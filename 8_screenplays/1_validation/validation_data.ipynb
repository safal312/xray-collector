#### Create Validation Data
This notebook cleans up the `clean_parsed_meta.json` to organize the data better and saves it to a csv file.

```python
import pandas as pd
import os
import json
import re
```

Clean parsed meta file is generated by the Movie Script Database repository after parsing the scripts.

```python
METADATA_DIR = "../../data/8_screenplays/"
```

```python
meta = ''
with open(f'{METADATA_DIR}/parsed_screenplays/metadata/clean_parsed_meta.json', 'r') as f:
  meta = json.load(f)
```

```python
rows = []

for key, value in meta.items():
  movie = {}

  if value.get('imdb'):
    movie['imdb_id'] = value['imdb']['id']
    movie['title'] = value['imdb']['title']

  if value.get('tmdb'): movie['tmdb_id'] = value['tmdb']['id']
  movie['filename'] = value['file']['file_name']

  if value.get('parsed'):
    movie['dialogue'] = value['parsed']['dialogue']
    movie['charinfo'] = value['parsed']['charinfo']
    movie['tagged'] = value['parsed']['tagged']

  movie['script_url'] = value['file']['script_url']

  rows.append(movie)
```

```python
df_meta = pd.DataFrame(rows, dtype=str)
```

```python
df_meta.head()
```

```python
# save this parsed metadata
min_meta = df_meta.drop(['tmdb_id', 'dialogue', 'charinfo', 'tagged', 'script_url'], axis=1)
```

```python
min_meta['match'] = 0
min_meta['alt_id'] = min_meta['imdb_id']
min_meta['script_url'] = df_meta['script_url']
min_meta['notes'] = ""
```
