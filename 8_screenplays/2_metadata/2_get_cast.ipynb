### Get Cast

This notebook extracts the cast for each movie and saves it into a file.

```python
import pandas as pd
from imdb import Cinemagoer
import json
import csv
import imdb
```

```python
ia = Cinemagoer()
```

```python
METADATA_DIR = "../../data/8_screenplays/2_metadata/"

df = pd.read_csv(f'{METADATA_DIR}/validated_movies.csv', dtype={'imdbID': str})
```

```python
df.head()
```

```python
df.shape
```

```python
df_temp = pd.DataFrame()

all_ids = df['imdbID']

for id in all_ids:
  movie = ia.get_movie(id)

  for cast in movie['cast']:
    roles = cast.currentRole
    # take = movie, cast
    if type(roles) == imdb.utils.RolesList:
      roles = ','.join([role['name'] for role in roles])
    else:
      if roles.get('name'):
        roles = roles['name']
      else: continue

    df_temp = pd.concat([df_temp, pd.DataFrame([[id, cast.personID, cast['name'], roles]])])
```

```python
df_temp.columns = ['imdbID', 'personID', 'person', 'characters']
```
