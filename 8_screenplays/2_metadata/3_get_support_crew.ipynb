This notebook gathers the support crew for all the movies.

```python
import pandas as pd
from imdb import Cinemagoer
import json
import csv
import imdb
import os
```

```python
ia = Cinemagoer()
```

```python
METADATA_DIR = "../../data/8_screenplays/2_metadata"
```

```python
df = pd.read_csv(f'{METADATA_DIR}/validated_movies.csv', dtype={'imdbID': str})
```

```python
df.shape
```

```python
cols = ['director', 'writer', 'producer', 'composer',
       'cinematographer', 'editor', 'editorial department', 'casting director',
       'production design', 'art direction', 'set decoration',
       'costume designer', 'make up', 'production manager',
       'assistant director', 'art department', 'sound crew', 'special effects',
       'visual effects', 'stunt performer', 'camera and electrical department',
       'casting department', 'costume department', 'location management',
       'music department', 'script department', 'transportation department',
       'miscellaneous crew', 'animation department', 'production department']
```

```python
fileexists = os.path.exists(f'{METADATA_DIR}/all_support_cast.csv')
mode = 'a' if fileexists else 'w'
```

```python
index = 0

if fileexists:
  df_temp = pd.read_csv(f'{METADATA_DIR}/all_support_cast.csv', dtype={'imdbID': str})
  index = len(df_temp['imdbID'].unique())
```

```python
take = ''     # for debugging

with open(f'{METADATA_DIR}/all_support_cast.csv', mode) as csvfile:
  writer = csv.writer(csvfile)
  if not fileexists: writer.writerow(['imdbID', 'role','personID', 'name'])
  for id in df['imdbID'][index:]:
    movie = ia.get_movie(id)
    rows = []

    for col in cols:

      take = movie, col
      res = movie.get(col)

      p_id = ''
      p_name = ''

      if type(res) != type(None):
        for person in res:
          if person.get('name'):
            p_name = person['name']

          if hasattr(person, 'personID'):
            if person.personID:
              p_id = person.personID
              rows.append([id, col, p_id, p_name])
          else:
            rows.append([id, col, p_id, p_name])
      else:
        rows.append([id, col, p_id, p_name])

    writer.writerows(rows)
```

```python
df = pd.read_csv(f'{METADATA_DIR}/all_support_cast.csv', dtype={'imdbID': str, 'personID': str})
```
