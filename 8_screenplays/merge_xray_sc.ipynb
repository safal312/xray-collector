### Merge Movies Metadata

Merge the movies metadata file with information on whether we have subtitles or screenplays for each of them.

```python
import pandas as pd
```

```python
df_x = pd.read_csv("../data/6_character_metadata/all_metadata_finalized.csv", dtype={"movie_id": str})
```

```python
df_x.columns
```

Read the file with metadata on screenplays

```python
df_sc = pd.read_csv("meta_imdb_and_stats.csv", dtype={"imdb": str})
```

```python
df_merged = df_x.merge(df_sc[["imdb", "filename"]], "left", left_on="movie_id", right_on="imdb")
```

```python
df_merged[~df_merged['imdb'].isnull()]
```

```python
df_merged['imdb'] = ~df_merged['imdb'].isnull()
df_merged = df_merged.rename(columns={"imdb": "screenplay", "filename": "screenplay_file"})
```

```python
df_merged[df_merged['screenplay']]
```

Read the file with metadata on subtitles collected.

```python
df_subtitles = pd.read_csv("../data/5_subtitles/subtitles_collected.csv")
```

```python
df_subtitles.head()
```

```python
df_merge_with_sub = df_merged.merge(df_subtitles[["file", "subtitle"]], "left", on="file")
```

```python
df_merge_with_sub.head()
```

```python
df_merge_with_sub.to_csv("../data/8_screenplays/metadata_with_screenplay_subtitles.csv", index=False)
```
