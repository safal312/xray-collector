### Merge Movies Metadata

Merge the movies metadata file with information on whether we have subtitles or screenplays for each of them.

```python
import pandas as pd
```

```python
df_x = pd.read_csv("../data/6_character_metadata/all_metadata_finalized.csv", dtype={"imdb_id": str})
```

```python
df_x.columns
```

Read the file with metadata on screenplays

```python
df_sc = pd.read_csv("../data/8_screenplays/3_character_matching/movies_with_high_coverage_and_matching/high_matches_meta.csv", dtype={"imdb_id": str})
```

```python
df_sc.columns
```

```python
df_merged = df_x.merge(df_sc[["imdb_id", "sp_file"]], how="left", on="imdb_id")
```

```python
df_merged[df_merged['imdb_id'].isnull()]
```

```python
# df_merged['imdb'] = ~df_merged['imdb'].isnull()
# df_merged = df_merged.rename(columns={"imdb": "screenplay", "filename": "screenplay_file"})
```

```python
df_merged['screenplay'] = ~df_merged.sp_file.isnull()
```

```python
df_merged[df_merged['screenplay']]
```

Read the file with metadata on subtitles collected.

```python
df_subtitles = pd.read_csv("../data/5_subtitles/subtitles_collected.csv")
```

```python
df_subtitles.shape
```

```python
df_subtitles.head()
```

```python
df_merge_with_sub = df_merged.merge(df_subtitles[["file", "subtitle"]], "left", on="file")
```

```python
df_merge_with_sub.head()
```

```python
df_merge_with_sub[['title', 'imdb_id', 'file', 'dir', 'synopsis', 'year', 'link', 'screenplay', 'sp_file', 'subtitle']].to_csv("../data/8_screenplays/metadata_with_screenplay_subtitles.csv", index=False)
```
